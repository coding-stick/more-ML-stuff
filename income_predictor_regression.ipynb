```python
df = pd.DataFrame(adult.data.targets['income'].map({'<=50K':0, '<=50K.':0, '>50K.':1, '>50K':1}))
df['age'] = adult.data.features['age']
df['education-num'] = adult.data.features['education-num']
df['hours-per-week'] = adult.data.features['hours-per-week']
```

```python
adult.data.features.columns
```

```python
dict_list = []
for i in adult.data.features.drop(columns=['age', 'fnlwgt', 'education-num', 'education', 'capital-gain', 'capital-loss', 'hours-per-week']).columns:
    dictionary = {"name": i}
    c = 1
    for k in adult.data.features[i]:
        if k not in dictionary: 
            dictionary[k] = c
            c+=1
    df[i] = adult.data.features[i].map(dictionary)
    dict_list.append(dictionary)
            
```

```python
final = df.dropna()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
X = final.drop(columns=['income'])
y = final['income']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)

LR = LogisticRegression().fit(X_train, y_train)

yhat = LR.predict(X_test)

print(f"Logistic regression: {accuracy_score(y_test, yhat)*100}%")
```

```python
education_map = {'Bachelors':13, 'HS-grad':9, '11th':7, 'Masters':14, '9th':5, 'Some-college':10,
       'Assoc-acdm':11, 'Assoc-voc':12, '7th-8th':4, 'Doctorate':16, 'Prof-school':15,
       '5th-6th':3, '10th':6, '1st-4th':2, 'Preschool':1, '12th':8}

def predict():
    
    i = ""
    age = int(input("Please enter your age: "))
    hpw = int(input("Please enter your hours per week: "))

    input_edu = ""
    while input_edu not in education_map:
        input_edu = input(f"Invalid education level, valid inputs are {', '.join(education_map.keys())}: ")
    print(f"Age: {age}, education level: {input_edu}, hours/week: {hpw}")
    inputs = [age, education_map[input_edu], hpw]

    for data in dict_list:
        while i not in data:
            i = input(f"Please enter {data['name']}, valid inputs are {", ".join(str(k) for k in data.keys())}. \n")
        print(f"{data['name']}: {i}")
        inputs.append(int(data[i]))
    
    print(f"\n    --> Your salary is {"ABOVE" if LR.predict([inputs])==1 else "BELOW"} 50K")

predict()  # Call the function to start the prediction process
```
